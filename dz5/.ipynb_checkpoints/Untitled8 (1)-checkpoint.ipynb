```r
install.packages("xts")
library(xts)
```

```r
dataPath <- './'
Data <- read.table(paste(dataPath,'stocks.csv',sep = '/'), header=TRUE,sep=',')
names(Data)[1] <- "DATE"
Data$DATE <- as.Date(Data$DATE, format = "%d.%m.%Y")
```

```r
head(Data)
```

```r
len<-  dim(Data)[1]
gazp <- diff(log(Data[,2]))
brent <- diff(log(Data[,3]))
copper <- diff(log(Data[,4]))
eurusd <- diff(log(Data[,5]))
gold <- diff(log(Data[,6]))
sp500 <- diff(log(Data[,7]))
usdrub <- diff(log(Data[,8]))

rates.xts <- as.xts(  cbind(gazp,brent, copper, eurusd, gold,sp500 , usdrub), order.by = Data[2:len,1])
```

```r
Data <- as.data.frame(rates.xts)
head(Data)
```

```r
model <-  lm(Data$gazp ~ Data$brent + Data$copper + Data$eurusd + Data$usdrub + Data$gold + Data$sp500 , data=Data)
summary(model)
```

p-value > 0.05 у всех предикторов, значит нет значимых предикторов. Также модель является вырожденной, потому что p-value F-statistic = 0.4469 > 0.05

Построим остатки

```r
plot(Data$gazp,type = "b",col='deepskyblue4',xlab='Time',main='gazp')
lines(model$fitted.values,col='firebrick1',lwd=3)
```

Построим гистограмму

```r
hist(model$residuals)
```

Проведем тест Шапиро-Вилка

```r
shapiro.test(model$residuals)
```

p-value получилось меньше 0.05, следовательно, гипотеза о нормальности отвергается.
